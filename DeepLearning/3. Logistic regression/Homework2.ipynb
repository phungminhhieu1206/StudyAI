- Hàm $ sigmoid(x) = \frac{1}{1 + e^{-x}}$

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

###Chứng minh hàm sigmoid được tìm ra để cho đạo hàm của loss function với w đẹp nhờ phép tính vi phân

- $\frac{\sigma(x)}{dx}$

$=\frac{dsigmoid(x)}{du} \frac{du}{dx}$ (chain rule, với $u = 1 + e^{-x}$)

$= \frac{e^{-x}}{(1+e^{-x})^2} = \frac{1}{(1+e^{-x})}*\frac{(1+e^{-x})-1}{(1+e^{-x})}$

$= sigmoid(x)(1-sigmoid(x))$


- Nhờ phép tính vi phân, hàm sigmoid được tìm ra để cho đạo hàm của Loss function với w đẹp. 

###Đạo hàm Loss function theo từng w

- $ln(u) = \frac{u'}{u}$ (*log trong bài toán này có thể xem như ln*)

$ $

- $\frac{dL}{d\hat{y}_i} = -\frac{d(y_i*log(\hat{y}_i) +(1-y_i)*log(1-\hat{y}_i))}{d\hat{y}_i} $

$= - (\frac{y_i}{\hat{y}_i}-\frac{1-y_i}{1-\hat{y}_i})$

$= \frac{\hat{y}_i- y_i}{\hat{y}_i(1-\hat{y}i)} $ (1)

$ $

- $\frac{d\hat{y}_i}{dw_0} = \frac{dsigmoid(w_0+w_1*x_1^{(i)}+w_2*x_2^{(i)})}{dw_0} $

$= \frac{dsigmoid(z)}{dz}\frac{dz}{dw_0}$ (với $z = w_0+w_1*x_1^{(i)}+w_2*x_2^{(i)}$) 

$= sigmoid(z)(1-sigmod(z))$ (với $\frac{dz}{dw_0}=1$)

$= \hat{y}_i*(1-\hat{y}_i)$  (2)

$ $

(1) (2):

$\frac{dL}{dw_0} = \frac{dL}{d\hat{y}}*\frac{d\hat{y}}{dw_0} = \frac{dL}{d\hat{y}}*\frac{d\hat{y}}{dz}*\frac{dz}{dw_2}= \hat{y}_i - y_i$

- Tương tự

$\frac{dL}{dw_1} = \frac{dL}{d\hat{y}}*\frac{d\hat{y}}{dz}*\frac{dz}{dw_1} = (\hat{y}_i - y_i)x_1^{(i)}$ (với $\frac{dz}{dw_1} = x_1^{(i)}$)

$\frac{dL}{dw_2} = \frac{dL}{d\hat{y}}*\frac{d\hat{y}}{dz}*\frac{dz}{dw_2} = (\hat{y}_i - y_i)x_1^{(i)}$ (với $\frac{dz}{dw_1} = x_2^{(i)}$)

###Code for ex1, ex2

```python
# Load data từ file csv
data = pd.read_csv('/content/drive/MyDrive/DeepLearning/3. Logistic regression/dataset.csv').values
N, d = data.shape
X = data[:, 0:d-1].reshape(-1, d-1)
Y = data[:, d-1].reshape(-1, 1)
```

```python
# Vẽ data bằng scatter
x_cho_vay = X[Y[:,0]==1]
x_tu_choi = X[Y[:,0]==0]

plt.scatter(x_cho_vay[:, 0], x_cho_vay[:, 1], c='red', edgecolors='none', s=30, label='cho vay')
plt.scatter(x_tu_choi[:, 0], x_tu_choi[:, 1], c='blue', edgecolors='none', s=30, label='từ chối')
plt.legend(loc=1)
plt.xlabel('mức lương (triệu)')
plt.ylabel('kinh nghiệm (năm)')
```

$X_{N,2} =
 \begin{pmatrix}
  1. & x_{1,2} \\
  1. & x_{2,2} \\
  \vdots & \vdots  \\
  1. & x_{n,2}
 \end{pmatrix}$

$W_{3,1} =
 \begin{pmatrix}
  0. \\
  0.1 \\
  0.1 
 \end{pmatrix}$

```python
# Thêm cột 1 vào dữ liệu x
X_bar = np.hstack((np.ones((N, 1)), X))
print(X_bar)
```

```python
# Khởi tạo tham số học
w = np.array([0.,0.1,0.1]).reshape(-1,1)
print(w)
```

$\hat{y} = sigmoid(XW)$
 
cost $L = - (y_i * log(\hat{y}_i) + (1-y_i) * log(1-\hat{y}_i))$
 
$w = w - lr * x^T(\hat{y} - y) $

```python
numOfIteration = 1000 #Số lần lặp bước 2
epoch = np.arange(1, numOfIteration + 1) #Trục epoch để vẽ đồ thị
#Khởi tạo các giá trị learning rate cần thử, và gán title
learning_rate = [{'name': 'Very low', 'value':0.001}, 
                 {'name': 'Medium', 'value':0.01},
                 {'name':'High', 'value':0.02}, 
                 {'name':'Very high', 'value':0.07}]
loss = np.zeros((len(learning_rate),numOfIteration)) #Loss với từng learning rate và số lần lặp (số epoch)

w1 = [w]*4 #thử nhiều giá trị learning_rate nên có nhiều w tương ứng
for j in range(len(learning_rate)):
  for i in range(1, numOfIteration):
    # Tính giá trị dự đoán
    y_predict = sigmoid(np.dot(X_bar, w1[j]))
    # Gradient descent np.dot: nhan ma tran, * / multiply : element-wise
    w1[j] = w1[j] - learning_rate[j]['value'] * np.dot(X_bar.T, y_predict - Y)	 
    loss[j][i] = -np.sum(np.multiply(Y, np.log(y_predict)) + np.multiply(1 - Y, np.log(1 - y_predict)))
    if (i == 1) | (i == numOfIteration // 2) | (i == numOfIteration-1):
      print(j, i, loss[j][i])
```

- Xác suất lớn hơn t (0<t<1) mới cho vay tiền

$ \hat{y}_i > t $

$<=> w_0 + w_1*x_1 + w_2*x_2 > - ln(\frac{1}{t} -1)$

- $x_2 = -\frac{w_0 + w_1*x_1 + ln(\frac{1}{t} -1)}{w_2}$

```python
# Vẽ đường phân cách.
t = 0.5
i_lr = np.argmin(loss[:,-1])
print(i_lr, w1[i_lr])
plt.scatter(x_cho_vay[:, 0], x_cho_vay[:, 1], c='red', edgecolors='none', s=30, label='cho vay')
plt.scatter(x_tu_choi[:, 0], x_tu_choi[:, 1], c='blue', edgecolors='none', s=30, label='từ chối')
plt.legend(loc=1)
plt.xlabel('mức lương (triệu)')
plt.ylabel('kinh nghiệm (năm)')

#plt.plot((4, 10),(-(w1[i_lr][0] + 4 * w1[i_lr][1] + np.log(1/t - 1)) / w1[i_lr][2], -(w1[i_lr][0] + 10 * w1[i_lr][1] + np.log(1/t - 1)) / w1[i_lr][2]), 'g') 
#(x1 start, x1 end) (x2 start, x2 end)
plt.plot((4, 10),(-(w1[0][0] + 4 * w1[0][1] + np.log(1/t - 1)) / w1[0][2], -(w1[0][0] + 10 * w1[0][1] + np.log(1/t - 1)) / w1[0][2]), 'black') 
plt.plot((4, 10),(-(w1[1][0] + 4 * w1[1][1] + np.log(1/t - 1)) / w1[1][2], -(w1[1][0] + 10 * w1[1][1] + np.log(1/t - 1)) / w1[1][2]), 'pink')
plt.plot((4, 10),(-(w1[2][0] + 4 * w1[2][1] + np.log(1/t - 1)) / w1[2][2], -(w1[2][0] + 10 * w1[2][1] + np.log(1/t - 1)) / w1[2][2]), 'r')
plt.plot((4, 10),(-(w1[3][0] + 4 * w1[3][1] + np.log(1/t - 1)) / w1[3][2], -(w1[3][0] + 10 * w1[3][1] + np.log(1/t - 1)) / w1[3][2]), 'y')
plt.show()
```

```python
A = np.array((0, 0, 1, 1)).reshape(-1,1)
B = np.array((0, 1, 0, 1)).reshape(-1,1)
C = np.logical_xor(A, B).astype(int)
data = np.hstack((A, B, C))

N, d = data.shape
x0 = data[:, 0:d-1].reshape(-1, d-1)
y = data[:, d-1].reshape(-1, 1)

x_1 = x0[y[:,0] == 1]
x_0 = x0[y[:,0] == 0]

#plt.scatter(x_1[:, 0], x_1[:, 1], c='red', edgecolors='none', s=30)
#plt.scatter(x_0[:, 0], x_0[:, 1], c='blue', edgecolors='none', s=30)
#plt.xlabel('X1')
#plt.ylabel('X2')
```

```python
x1 = x0[:,0].reshape(-1,1)
x2 = x0[:,1].reshape(-1,1)
x = np.hstack((np.ones((N, 1)), x1, x2, x1**2, x2**2, x1*x2))
w = np.array([-5.,1.,1.,1.,1.,-6.]).reshape(-1,1)
```

```python
numOfIteration = 1000
epoc = np.arange(1, numOfIteration + 1)

learning_rate = [{'name': 'Very low', 'value':0.0001}, 
                 {'name': 'Medium', 'value':0.001},
                 {'name':'High', 'value':0.01}, 
                 {'name':'Very high', 'value':0.1}]
cost = np.zeros((len(learning_rate),numOfIteration))

w1 = [w] * 4 #thử nhiều giá trị learning_rate nên có nhiều w tương ứng
for j in range(len(learning_rate)):
  for i in range(1, numOfIteration):
    y_predict = sigmoid(np.dot(x, w1[j]))
    # Gradient descent np.dot: nhan ma tran, * / multiply : element-wise
    w1[j] = w1[j] - learning_rate[j]['value'] * np.dot(x.T, y_predict - y)	 
    cost[j][i] = -np.sum(np.multiply(y, np.log(y_predict)) + np.multiply(1 - y, np.log(1 - y_predict)))
    if (i == 1) | (i == numOfIteration // 2) | (i == numOfIteration-1):
      print(j, i, cost[j][i])

limy = cost[0][1] *1.5
plt.figure(figsize=(6, 6))
plt.plot(epoc, cost[0], label = learning_rate[0]['name'])
plt.plot(epoc, cost[1], label = learning_rate[1]['name'])
plt.plot(epoc, cost[2], label = learning_rate[2]['name'])
plt.plot(epoc, cost[3], label = learning_rate[3]['name'])
plt.legend()
plt.title('Loss function by Epoch and Learning Rate')
```

```python
# Vẽ đường phân cách.
t = 0.5
i_lr = np.argmin(cost[:,-1])
print(i_lr, w1[i_lr])
plt.scatter(x_1[:, 0], x_1[:, 1], c='red', edgecolors='none', s=30)
plt.scatter(x_0[:, 0], x_0[:, 1], c='blue', edgecolors='none', s=30)
plt.xlabel('X1')
plt.ylabel('X2')

x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
x, y = np.meshgrid(x, y)
plt.contour(x, y, w1[3][0] + w1[3][1]*x + w1[3][2]*y + w1[3][3]*x**2 + w1[3][4]*y**2 + w1[3][5]*x*y + t, [0], c='k')

plt.show()
```

```python
import sys
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import PolynomialFeatures
```

```python
X, y = [], []
for i in range(2):
    for j in range(2):
        X.append([i, j])
        y.append(i ^ j)
 
for x in X:
    x.extend([x[0]*x[0], x[1]*x[1], x[0]*x[1]])
 
X = np.array(X)
y = np.array(y)
 

clf = Perceptron(max_iter = 10, shuffle=False).fit(X, y)
if clf.score(X, y) != 1.0:
    print('Failed to fit the data.')
    sys.exit(1)
 
plt.title("%fx%+fy%+fx^2%+fy^2%+fxy%+f" % (clf.coef_[0, 0], clf.coef_[0, 1],
            clf.coef_[0, 2], clf.coef_[0, 3], clf.coef_[0, 4],
            clf.intercept_))
 
for i in range(len(y)):
    if y[i] == 1:
        plt.scatter(X[i, 0], X[i, 1], c='r')
    elif y[i] == 0:
        plt.scatter(X[i, 0], X[i, 1], c='b')
 
XX, YY = np.mgrid[-2:3:200j, -2:3:200j]
XXX = []
for xs, ys in zip(XX, YY):
    for x_, y_ in zip(xs, ys):
        XXX.append([x_, y_, x_*x_, y_*y_, x_*y_])
Z = clf.decision_function(XXX)
 
Z = Z.reshape(XX.shape)
plt.contour(XX, YY, Z, levels=[0])
plt.show()
```

```python
X, y = [], []
for i in range(2):
    for j in range(2):
        X.append([i, j])
        y.append(i ^ j)
 
X = PolynomialFeatures(interaction_only=True).fit_transform(X)
clf = Perceptron(fit_intercept=False, max_iter=15, shuffle=False).fit(X, y)
 
if clf.score(X, y) != 1.0:
    print('Failed to fit the data.')
    sys.exit(1)
 
plt.title("%fx%+fy%+fxy%+f" % (clf.coef_[0, 1], clf.coef_[0, 2],
            clf.coef_[0, 3], clf.coef_[0, 0]))
 
for i in range(len(y)):
    if y[i] == 1:
        plt.scatter(X[i, 1], X[i, 2], c = 'r')
    elif y[i] == 0:
        plt.scatter(X[i, 1], X[i, 2], c = 'b')
 
XX, YY = np.mgrid[-2:3:200j, -2:3:200j]
XXX = []
for xs, ys in zip(XX, YY):
    for x_, y_ in zip(xs, ys):
        XXX.append([1, x_, y_, x_*y_])
Z = clf.decision_function(XXX)
 
Z = Z.reshape(XX.shape)
plt.contour(XX, YY, Z, levels=[0])
plt.show()
```

```python
import pandas as pd
# load dataset
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
pima = pd.read_csv('/content/drive/MyDrive/DeepLearning/3. Logistic regression/diabetes.csv', header=0, names=col_names)
```

```python
pima.head()
```

- Chọn lọc các column ra 2 dạng, đặc trưng cho biến kết quả (target variable) và features 

```python
#split dataset in features and target variable
feature_cols = ['pregnant', 'insulin', 'bmi', 'age','glucose','bp','pedigree']
X = pima[feature_cols] # Features
y = pima.label # Target variable
```

- Tách data thành 2 tập: training set và test set
- Sử dụng hàm train_test_split của thư viện sklearn. Cần 3 tham số về feature, target, và test_size. Có thể sử dụng random_state để chọn các record ngẫu nhiên.
- test_size 0.25 chia data thành 75% training, 25% test. 

```python
# split X and y into training and testing sets
#from sklearn.cross_validation import train_test_split, không import được
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

- Dự đoán và phát triển model
- Import model Logistic Reg. và tạo một classifier object tương ứng. 
- Fit bộ training vào và chạy dự đoán trên bộ test.

```python
# import the class
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression()

# fit the model with data
logreg.fit(X_train,y_train)

#
y_pred=logreg.predict(X_test)
```

- Đánh giá model bằng Confusion Matrix.
- CM được dùng để đánh giá hoạt động của classification model, và hình dung hiệu suất của 1 thuật toán.
- CM về cơ bản là số lượng các dự đoán đúng/sai được tổng hợp thành từng lớp.

```python
# import the metrics class
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

- Kích thước 2x2 vì đây là phân loại nhị phân (binary classification). Hai lớp 0 và 1
- Giá trị đường chéo đại diện cho các dự đoán chính xác. Và các giá trị còn lại là dự đoán chưa chính xác.

- Biểu diễn CM bằng Heatmap

```python
# import required modules
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

- Đánh giá CM bằng 3 thông số


```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```

- Tỷ lệ phân loại là 80% được coi là độ chính xác tốt.
- $Precision = \frac{True Positive}{(True Positive + True Negative)}$

$=> Precision = \frac{Có bệnh Chính xác}{(Có bệnh Chính xác + Có bệnh Không chính xác)}$
- $Recall = \frac{True Positive}{(True Positive + False Negative)}$

$=> Recall = \frac{Có bệnh Chính xác}{(Có bệnh Chính xác + Không bệnh Không chính xác)}$. 
- Precision và Recall được chú trọng để cân nhắc khi dự đoán có thể gây ra hậu quả nghiêm trọng (vd dự đoán email Spam nhưng không phải Spam (Precision), dự đoán Không bệnh nhưng bệnh nhân Có bệnh (Recall))
- Có thể dùng tham số F-1 Score để dung hòa:
$ F1 = 2\frac{Precision*Recall}{Precision+Recall}$

- F1 càng cao càng tốt, lý tưởng nhất là = 1

- ROC (Receiver Operating Characteristic)
- Đồ thị của tỉ lệ dương tính thực so với dương tính giả.

```python
y_pred_proba = logreg.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

- auc là 0.86. auc = 1 là tốt, 0.5 là vô giá trị

-Nhược điểm: Không thể xử lý số lượng lớn features/variables. Không hoạt động tốt khi các features không tương quan với target variable, hay các features rất giống nhau hoặc có tương quan với nhau.
